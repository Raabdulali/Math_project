```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.fft import fft, fftfreq
import random
```

```python
#initial parameters/random parameters
N = 1000
T = 1.0 / 5000
t = random.randint(10,150)
h = random.randint(10,150)
u = random.randint(10,150)



#creating 3 sine waves
x = np.linspace(0.0, N*T, N, endpoint=False)
y = np.sin(t * 2.0*np.pi*x) 
z = 0.5*np.sin(h * 2.0*np.pi*x)
l = 1.5*np.sin(u * 2.0*np.pi*x)


#combining all sine waves
c = np.sin(t * 2.0*np.pi*x) + 0.5*np.sin(h * 2.0*np.pi*x) + 1.5*np.sin(u * 2.0*np.pi*x)


#applying the fourier transform
yf = fft(c)
xf = fftfreq(N, T)[:N//2]



#plot final sine wave
plt.plot(x,c)
plt.show()


#plot fourier transform
plt.plot(xf, 2.0/N * np.abs(yf[0:N//2]))
plt.xlim([0, 170])
plt.grid()
plt.show()
```

```python
#check values from fourier tranform
print('t = ', t)
print('h = ', h)
print('u = ', u)

#print the 3 individual waves for kicks
plt.plot(x,y)
plt.show()
plt.plot(x,z)
plt.show()
plt.plot(x,l)
plt.show()
```
